```python
import cv2
import numpy as np
import math
from scipy.ndimage.filters import gaussian_filter, maximum_filter, sobel
from myfiltering import cross_correlation_2d, convolve_2d, gaussian_blur_kernel_2d, sobel_kernel, sobel_image


def rgb2gray(image):
    red, green, blue = image[:,:,0], image[:,:,1], image[:,:,2]
    return (0.2125 * red + 0.7154 * green + 0.0721 * blue) / 255.

def compute_corner_response(image, gaussian_sigma, harris_corner_k):
        # TODO: Compute the Harris corner response
        R = np.zeros_like(image)
        
        #sobel kernel
        
        Ix = sobel(image, axis=1)
        Iy = sobel(image, axis=0)
        
        '''''Ix = sobel(image, axis=1)
        Iy = sobel(image, axis=0)
        
        Ix = (Ix-np.max(Ix))/(np.max(Ix)-np.min(Ix))
        Iy = (Iy-np.max(Iy))/(np.max(Iy)-np.min(Iy))
        
        Ix = (abs(Ix*255)).astype(np.uint8)
        Iy = (abs(Iy*255)).astype(np.uint8)'''

        
        Ix_2 = Ix**2
        Ix_Iy = Ix * Iy
        Iy_2 = Iy**2
        
        
        #window_size and gaussian filter(used for weighting)
        win_size = 7
        half_width = math.floor(win_size/2)
        #gaussian = gaussian_blur_kernel_2d(win_size, gaussian_sigma)
        
        
        ''''
        Ix_2 = convolve_2d(Ix_2, gaussian, 'valid')
        Ix_Iy = convolve_2d(Ix_Iy, gaussian, 'valid')
        Iy_2 =convolve_2d(Iy_2, gaussian, 'valid')
        '''       
        #cv2.imshow("Ix_2afte gaussian", Ix_2)
        #cv2.waitKey()
        '''
        new_Ix_2 = np.zeros_like(image)
        new_Ix_2[half_width:-half_width, half_width:-half_width] = Ix_2
        new_Ix_Iy = np.zeros_like(image)
        new_Ix_Iy[half_width:-half_width, half_width:-half_width] = Ix_Iy
        new_Iy_2  = np.zeros_like(image)
        new_Iy_2[half_width:-half_width, half_width:-half_width] = Iy_2
        '''

        #get row and col for for-loop
        row, col = R.shape
        
        #initialize k
        k = harris_corner_k
        
        
        
        #get R(x,y)
        print("start")
        for i in range(half_width, row-half_width):
            for j in range(half_width, col-half_width):
                
                A = np.sum(gaussian_filter(Ix_2[i-half_width:i+half_width+1, j-half_width:j+half_width+1], gaussian_sigma))
                B = np.sum(gaussian_filter(Ix_Iy[i-half_width:i+half_width+1, j-half_width:j+half_width+1], gaussian_sigma))
                C = np.sum(gaussian_filter(Iy_2[i-half_width:i+half_width+1, j-half_width:j+half_width+1], gaussian_sigma))
                M = np.array([[A,B],[B,C]])
                #M = np.array([[new_Ix_2[i,j],new_Ix_Iy[i,j]],[new_Ix_Iy[i,j],new_Iy_2[i,j]]])
                det = np.linalg.det(M)
                trace = np.trace(M)
                R[i,j] = det - k*(trace**2)
        
        
        
        #R = (Ix_2 * Iy_2 / (Ix_Iy**2)) - k*((Ix_2 + Iy_2)**2)

        return R
    
def get_keypoints(R, maxfilter_window_size, max_num_features):
        # TODO: apply non-maximum suppression and obtain the up-to-K strongest
        # features having R(x,y) > 0
        
        w = maxfilter_window_size
                
        local_max = maximum_filter(R,size=(w,w))
        local_max[local_max != R] = 0
        
       
        
        reverse_sort = np.argsort(local_max.ravel())
        sort  = reverse_sort[::-1]
        K = max_num_features
        
        #to get (x,y) and make the row into y and col index into x
        y_coor, x_coor = np.unravel_index(sort[:K], R.shape)
        
        
        coor = np.column_stack((x_coor,y_coor))
        
        
        return local_max, coor
    
```

```python
def call(image1, keypoints1, image2, keypoints2, window_size):
        d1 = get_descriptors(image1, keypoints1, window_size)
        d2 = get_descriptors(image2, keypoints2, window_size)
        

        match_matrix =  match_ssd(d1, d2)#use ssd
        #match_matrix = match_ncc(d1, d2)
        matches = compute_matches(match_matrix)

        return matches
    

def get_descriptors(image, keypoints, window_size):
        # TODO: Obtain descriptors for the provided keypoints
        
        #define output
        output = []
        
        #define N, the number of keypoints
        N = keypoints.shape[0]
        
        
        #define half window size
        half_width = math.floor(window_size/2)
        
        #zero padding
        #zero_matrix = np.zeros((self.window_size,self.window_size))
        #zero_matrix[half_width] = 1
        #new_im = convolve_2d(image, zero_matrix, 'full')

        #get x-coordinate and y-coordinate in the key point and their descriptor in ravelled form
        for coor in keypoints:
            x_coor = coor[0]
            y_coor = coor[1]
            output.append(image[y_coor - half_width:y_coor + half_width+1, x_coor - half_width:x_coor + half_width+1])
          
        output = np.array(output)
        
        output = output.reshape((N, -1))
         
            
        return output
    
def match_ssd(d1, d2):
        # TODO: implement SSD comparison of the two descriptor sets
        output = np.zeros((d1.shape[0], d2.shape[0]))
        for d_1 in range(d1.shape[0]):
            for d_2 in range(d2.shape[0]):
                output[d_1,d_2] = np.sum((d1[d_1] - d2[d_2])**2)
        return output
    
    
def match_ncc(d1, d2):
        # TODO: implement 1 - NCC comparison of the two descriptor sets
        ncc_output = np.zeros((d1.shape[0], d2.shape[0]))
        for d_1 in range(d1.shape[0]):
            for d_2 in range(d2.shape[0]):
                normalized_d1 = (d1[d_1] - np.mean(d1[d_1])) / np.std(d1[d_1])
                normalized_d2 = (d2[d_2] - np.mean(d2[d_2])) / np.std(d2[d_2])
                ncc_output[d_1,d_2] = 1 - np.sum(normalized_d1 * normalized_d2) #this is 1-ncc
        return ncc_output
    
def compute_matches(match_matrix):
        # TODO: implement one-to-one matching for the match matrix
        matching_output = []
        for row in range(match_matrix.shape[0]):
            for col in range(match_matrix.shape[1]):
                if (col == np.argmin(match_matrix[row,:])) & (row == np.argmin(match_matrix[:,col])):
                    arr = np.array([row, col])
                    matching_output.append(arr)
                    
        matching_output = np.array(matching_output)
        matching_output = matching_output.reshape((-1, 2))
        return matching_output


matches_1_and_2 = call(image, key_coordinates, image2, key_coordinates_2, 7)
```

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    # Python 2
    from itertools import izip
except ImportError:
    # Python 3
    izip = zip

#-------------------------------------------------------------------------------

# display keypoints in an image
#
# inputs:
# - image: input image, assumed to be grayscale
# - keypoints: Nx2 array of keypoint (x,y) pixel locations in the image
# returns: None
def plot_keypoints(image, keypoints):
    plt.imshow(image, cmap="gray")
    plt.scatter(keypoints[:,0], keypoints[:,1], s=8, c="r")
    plt.xlim(0, image.shape[1])
    plt.ylim(image.shape[0], 0)
    plt.savefig("myfig.jpg")
    
def plot_matches(image1, keypoints1, image2, keypoints2, matches):
    keypoints1 = keypoints1[matches[:, 0]]
    keypoints2 = keypoints2[matches[:, 1]]

    # show the original images side-by-side
    im = np.column_stack((image1, image2))
    plt.imshow(im, cmap="gray")
    plt.scatter(keypoints1[:, 0], keypoints1[:, 1],
		  facecolors='none', edgecolors='r')
    plt.scatter(keypoints2[:, 0] + image1.shape[1], keypoints2[:, 1],
		  facecolors='none', edgecolors='r')

    # draw the matches with random colors
    for kp1, kp2 in izip(keypoints1, keypoints2):
        x = (kp1[0], kp2[0] + image1.shape[1])
        y = (kp1[1], kp2[1])
        c = (2. * np.random.rand(3) + 1.) / 3. # RGB \in (0.5, 1)
        plt.plot(x, y, color=c, linewidth=1.0)

    plt.gca().axis('off')
    plt.tight_layout()
    #plt.savefig("myfig.jpg")
    

a =  plot_keypoints(image, key_coordinates)
#a = plot_matches(image, key_coordinates, image2, key_coordinates_2, matches_1_and_2)
```
